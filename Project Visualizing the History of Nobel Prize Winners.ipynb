The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
```

```python
df=pd.read_csv('data/nobel.csv')
type(df)
```

```python
df.head()
```

```python
top_gender = df.value_counts('sex')
top_gender = top_gender.index[0]
top_gender = str(top_gender)
top_gender
```

```python
top_country = df.value_counts('birth_country').index[0]
top_country
```

```python
df1 = df[['year','birth_country']]
df1 = df1[df1['birth_country'] == 'United States of America']
df1.head()
```

```python
df1['decade'] = (df1['year'] // 10) * 10
df1 = df1.groupby('decade')['birth_country'].count()
df1.head()
```

```python
df2=pd.DataFrame(df1)
df2['contribution'] = round(df2['birth_country']/df2['birth_country'].sum()*100,2)
df2 = df2.sort_values('contribution', ascending=False)
max_decade_usa = int(df2.index[0])
max_decade_usa
```

```python
df['is_female'] = df['sex'] == 'Female'
df['decade'] = (df['year'] // 10) * 10
```

```python
df3 = df.groupby(['decade', 'category'], as_index=False)['is_female'].mean()
df3.head()
```

```python
max_mean = df3['is_female'].max()
df3 = df3[df3['is_female'] == max_mean]
df3 = df3[['decade', 'category']]
df3
```

```python
max_female_dict = {df3.iloc[0,0]:df3.iloc[0,1]}
max_female_dict
```
